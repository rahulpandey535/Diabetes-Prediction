```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#saparating data 
x=d_data.drop(columns='Outcome',axis=1)
y=d_data['Outcome']
```

```python
print (x)
```

```python
print(y)
```

```python
#data standerization

scaler=StandardScaler()
```

```python
std_data=scaler.fit_transform(x)
```

```python
print(std_data)
```

```python
x=std_data
y=d_data['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
#train test split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#training the model

classifier=svm.SVC(kernel='linear')
```

```python
#training svm classifier
classifier.fit(x_train,y_train)
```

```python
#model evaluation
#acc score

x_train_pre=classifier.predict(x_train)
training_data_acc=accuracy_score(x_train_pre,y_train)
```

```python
print('Accuracy score of the training data:',training_data_acc)
```

```python
x_test_pre=classifier.predict(x_test)
test_data_acc=accuracy_score(x_test_pre,y_test)
```

```python
print('Accuracy score of the test data:',test_data_acc)
```
